```python
x_val_vect = []

for i in val_yelp['photo_id']:
    a= int(i)
    print ("reading image:"+str(a) + ".jpg")
    img = ('D:/Share_Folder/Pace_3rd_Sem/Big_Data_Analytics/Project Files/train_photos/'  + str(a) + ".jpg")
    img = cv2.imread(img)
    img = cv2.resize(img,(100,100))
    img = img.transpose((2,0,1))
    # img = io.imread(img)
    # img = resize(img,(100,100))
    # img = img.transpose()
    x_val_vect.append(img)

x_val_vect = np.array(x_val_vect)

f = h5py.File("valyelp.h5")
f['x'] = x_val_vect
f['y'] = y_val_vect
f.close()
```
